```python
import numpy as np
import pandas as pd
df=pd.read_csv("/content/lung_cancer_examples.csv")
df
```

```python
df.columns
```

```python
x=df.iloc[:,2:-1]
y=df.iloc[:,-1]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
y_test
```

```python
#normalisation
#standard scaler= converting all values into same type  z=(x-u)/s  u=mean of training data, s=std deviation
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
```

```python
#model creation
from sklearn.svm import SVC # support vector classifier
model=SVC()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics._plot.confusion_matrix import confusion_matrix
from sklearn.metrics import classification_report,accuracy_score,ConfusionMatrixDisplay
labels=["0","1"]
result=confusion_matrix(y_test,y_pred)
cm=ConfusionMatrixDisplay(result,display_labels=labels)
score=accuracy_score(y_test,y_pred)
print(result)
print(score) 
report=classification_report(y_pred,y_test)
print(report)
```
